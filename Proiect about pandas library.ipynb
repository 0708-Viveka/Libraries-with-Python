```python
import pandas as pd


titanic_df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")


print(titanic_df.head())


print(titanic_df.info())

# Display basic statistics of the dataset
print(titanic_df.describe())
```

```python

print(titanic_df.isnull().sum())
```

```python
# Fill missing values in the 'Age' column with the median age
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Fill missing values in the 'Embarked' column with the most frequent value
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)

# Drop rows where the 'Cabin' column is missing
titanic_df.dropna(subset=['Cabin'], inplace=True)

# Verify that there are no missing values left
print(titanic_df.isnull().sum())
```

```python
# Calculate survival rate
survival_rate = titanic_df['Survived'].mean()
print(f'Survival rate: {survival_rate:.2f}')

# Survival rate by gender
survival_rate_by_gender = titanic_df.groupby('Sex')['Survived'].mean()
print(survival_rate_by_gender)

# Survival rate by passenger class
survival_rate_by_class = titanic_df.groupby('Pclass')['Survived'].mean()
print(survival_rate_by_class)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot survival rate by gender
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.show()

# Plot survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=titanic_df)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Plot age distribution
plt.hist(titanic_df['Age'], bins=30, edgecolor='k')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create a new feature 'FamilySize' from 'SibSp' and 'Parch'
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1

# Create a new feature 'IsAlone'
titanic_df['IsAlone'] = (titanic_df['FamilySize'] == 1).astype(int)

# Create a new feature 'Title' from the 'Name' column
titanic_df['Title'] = titanic_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Display the first few rows to see the new features
print(titanic_df.head())
```

```python
# Convert 'Sex' column to numerical values
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1}).astype(int)

# Convert 'Embarked' column to numerical values using one-hot encoding
titanic_df = pd.get_dummies(titanic_df, columns=['Embarked'], drop_first=True)

# Convert 'Title' column to numerical values using one-hot encoding
titanic_df = pd.get_dummies(titanic_df, columns=['Title'], drop_first=True)

# Display the first few rows to see the encoded features
print(titanic_df.head())
```
