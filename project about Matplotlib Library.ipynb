```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter plot")
plt.xlabel("month")
plt.ylabel("numers")

c = ['red', 'magenta', 'green', 'orange', 'blue', 'brown']

plt.scatter(x,y, color = c, s = 200)
plt.show()
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm, data = df1, hue = "sex"  )
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"] 
sns.barplot(x ="island", y ="bill_length_mm", data = df1, hue = "sex", order = order1)
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"] 
sns.barplot(x ="island", y ="bill_length_mm", data = df1, hue = "sex", order = order1, hue_order = ['Female', 'Male'])
plt.show()
```

```python
order1 = ["Dream", "Torgersen", "Biscoe"] 
sns.barplot(x ="island", y ="bill_length_mm", data = df1, color = "green")
plt.show()
```

```python
sns.barplot(x ="island", y ="bill_length_mm", data = df1, hue = "sex", order = order1, palette = "Accent")
plt.show()
```

```python
sns.barplot(x ="island", y ="bill_length_mm", data = df1, hue = "sex", order = order1,
            palette = "gist_heat_r", saturation = 0.7)
plt.show()
```
